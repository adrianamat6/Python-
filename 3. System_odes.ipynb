```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def mixer(x,t,Tf,Caf):

    Ca = x[0]
    T = x[1]

    #Parameter
    q =100
    V = 100

    dCadt = q * (Caf - Ca)/V
    
    dTdt = q * (Tf - T)/V
    
    return [dCadt, dTdt]

#Initial condition
Ca0 = 0.0
T0 = 350.0

#Feed temperature (K)
Tf = 300

#Feed concentration (mol/L)
Caf = 1

# Time interval
t = np.linspace(0,10,100)

# Simulate mixer
y = odeint(mixer,[Ca0, T0],t,args=(Tf,Caf))

Ca = y[:,0]
T = y[:,1]

plt.subplot(1,2,1)
plt.plot(t,Ca)
plt.xlabel('Time (h)')
plt.ylabel('Ca (mol/s)')

plt.subplot(1,2,2)
plt.plot(t,T)
plt.xlabel('Time (h)')
plt.ylabel('T (K))')

```

```python
#Save data
data = np.vstack((t,Ca.T,T.T))
data = data.T
np.savetxt(' data.txt',data,delimiter=',')
```
